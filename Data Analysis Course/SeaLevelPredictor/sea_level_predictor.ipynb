```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    df = pd.read_csv("epa-sea-level.csv")
    fig, ax = plt.subplots(figsize=(18,6))
    plt.scatter(data=df,x="Year",y="CSIRO Adjusted Sea Level")
    plt.xlabel("Year")
    plt.ylabel("Sea Level(inches)")
    plt.title("Rise in Sea Level")
    linr1 = linregress(df["Year"],df["CSIRO Adjusted Sea Level"],)
    xA = np.arange(df['Year'].min(),2051,1)
    yA = xA*linr1.slope + linr1.intercept
    plt.plot(xA,yA)
    df2000 = df[df["Year"]>=2000]
    linr2 = linregress(df2000["Year"],df2000["CSIRO Adjusted Sea Level"])
    xB = np.arange(df2000['Year'].min(),2051,1)
    yB = xB*linr2.slope + linr2.intercept
    plt.plot(xB,yB)
    plt.savefig('sea_level_plot.png')
    return plt.gca()


```

```python
draw_plot()
```
