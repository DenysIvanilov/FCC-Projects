```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("medical_examination.csv")
```

```python
df["overweight"] = (df["weight"] / pow((df["height"] / 100), 2)).apply(lambda x: 1 if x > 25 else 0)
df["cholesterol"] = df["cholesterol"].apply(lambda x: 1 if x > 1 else 0)
df["gluc"] = df["gluc"].apply(lambda x: 1 if x > 1 else 0)
```

```python
def draw_cat_plot():
    df_cat = pd.melt(df, id_vars="cardio", var_name="variable", value_vars=["active","alco","cholesterol","gluc","overweight","smoke"])
    fig = sns.catplot(data=df_cat, col="cardio", kind="count", x="variable", hue="value")
    
    fig.savefig('catplot.png')
    return fig
```
